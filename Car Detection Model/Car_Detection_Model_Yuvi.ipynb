```python
# ==========================================================
# ğŸš€ Training YOLOv8 on Your Car Detection Dataset
# ==========================================================

# 1ï¸âƒ£ Install YOLOv8 (Ultralytics library)
!pip install ultralytics --quiet

# 2ï¸âƒ£ Import YOLO
from ultralytics import YOLO

# 3ï¸âƒ£ Initialize model with pretrained weights
# 'yolov8n.pt' = nano model (fastest, good for small datasets)
model = YOLO("yolov8n.pt")

# 4ï¸âƒ£ Train the model
# You can raise epochs to 50â€“100 if you have GPU time
model.train(
    data="/content/Car-Detection-Model-1/data.yaml",  # auto-generated YAML from Roboflow
    epochs=30,
    imgsz=640,
    batch=8,
    name="car_detection_model_v1"
)

# 5ï¸âƒ£ Evaluate model performance after training
model.val()

print("âœ… Training complete! Check /content/runs/detect/car_detection_model_v1/ for results.")
```
