```python
# ==========================================================
# 🚀 Training YOLOv8 on Your Car Detection Dataset
# ==========================================================

# 1️⃣ Install YOLOv8 (Ultralytics library)
!pip install ultralytics --quiet

# 2️⃣ Import YOLO
from ultralytics import YOLO

# 3️⃣ Initialize model with pretrained weights
# 'yolov8n.pt' = nano model (fastest, good for small datasets)
model = YOLO("yolov8n.pt")

# 4️⃣ Train the model
# You can raise epochs to 50–100 if you have GPU time
model.train(
    data="/content/Car-Detection-Model-1/data.yaml",  # auto-generated YAML from Roboflow
    epochs=30,
    imgsz=640,
    batch=8,
    name="car_detection_model_v1"
)

# 5️⃣ Evaluate model performance after training
model.val()

print("✅ Training complete! Check /content/runs/detect/car_detection_model_v1/ for results.")
```

```python
# ==========================================================
# 🎬 Run Inference on a Video
# ==========================================================

from google.colab import files

# 1️⃣ Upload a video file
uploaded = files.upload()
video_path = list(uploaded.keys())[0]

# 2️⃣ Run detection on video
model.predict(source=video_path, save=True, conf=0.4, stream=False)

print("✅ Video processed successfully! Check the 'runs/detect/predict' folder for output video.")

```
